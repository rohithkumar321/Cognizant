```python

print(df['category'].value_counts())
```

```python
total_sales = df['total'].sum()
print('Total Sales:',total_sales)
```

```python
average_unit_price = df['unit_price'].mean()
print('Average Unit Price:', average_unit_price)
```

```python
total_quantity = df['quantity'].sum()
print('Total Quantity Sold:', total_quantity)
```

```python
df['timestamp']=pd.to_datetime(df['timestamp'])
df['purchase_hour'] = df['timestamp'].dt.hour
df['purchase_hour']
```

```python
hourly_sales = df.groupby('purchase_hour').size()
hourly_sales
```

```python
most_selling_hour = hourly_sales.idxmax()
most_sales_count = hourly_sales.max()

print("\nMost Selling Hour:", most_selling_hour, "with", most_sales_count, "sales")
```

```python
cash_transactions = df[df['payment_type'] == 'cash']
cash_item_counts = cash_transactions['category'].value_counts()
cash_item_counts.nlargest(4)
```

```python
sales_by_customer_type = df.groupby('customer_type')['total'].sum()
sales_by_customer_type.nlargest(2)
```

```python
product_sales = df['product_id'].value_counts()
print("product id",product_sales.idxmax(),"No.of sales",product_sales.max())


```

```python
# Check for missing values
print(df.isnull().sum())


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['total'], kde=True)
plt.title('Distribution of Total Sales')
plt.xlabel('Total')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='customer_type', y='total', data=df)
plt.title('Total Sales by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Total Sales')
plt.show()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.set_index('timestamp')
df.resample('D')['total'].sum().plot(kind='line', figsize=(10, 6))
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='payment_type', data=df)
plt.title('Distribution of Payment Types')
plt.xlabel('Payment Type')
plt.ylabel('Count')
plt.show()
```
